```python
dfGOV = pd.read_csv("https://raw.githubusercontent.com/DerikVoid/Projeto-metadado_Gov/refs/heads/main/comprasGOV-diario-VW_FT_PNCP_COMPRA_ITEM-latest.csv")
dfGOV.head(5)
```

```python
dfGOV.info()
```

```python
#Filtra todas as colunas que não possuem valores algum dentro delas
dfGOV = dfGOV.iloc[:, 0: 136].dropna(how='all', axis=1)
dfGOV.head()
```

```python
#Limpando as colunas especificas que possuem muitos poucos dados
dfGOV_limps = dfGOV.drop(columns = ['codigo_classe', 'descricao_detalhada', 'cod_item_catalogo', 'TXT_LINK_MATERIAL', 'TXT_LINK_CLASSE'])
dfGOV = dfGOV_limps
dfGOV.head()
```

```python
#Organizando as Colunas


#Dropando colunas disnecessárias
dfGOV_limps = dfGOV_limps.drop(columns='material_ou_servico').rename(
    columns={'material_ou_servico_nome' : 'material_ou_servico' ,
    'srk_pncp_item_compra':'id_pncp_item_compra'})




dfGOV_limps.head()
```

```python
# salvar o DataFrame correto
dados_limpsG = dfGOV_limps
dados_limpsG.to_csv('dadosGOV_tratados.csv', index=False)

# baixar
from google.colab import files
files.download('dadosGOV_tratados.csv')
```
